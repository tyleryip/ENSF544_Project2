```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import rand
from sklearn import preprocessing
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

%matplotlib inline

source_file = "winequality-white.csv"
data = pd.read_csv(source_file, delimiter=";")

label = data['quality']
features = data.loc[:, data.columns != 'quality']

#Standardize data
standard = preprocessing.scale(features)

#Split
x_train, x_test, y_train, y_test = train_test_split(standard, label, random_state=1)

x_train, x_validate, y_train, y_validate = train_test_split(x_train, y_train, random_state=1, test_size = 0.25)

# creating, training and plotting regression
regressor = SVR(kernel='rbf', C=0.95, gamma=0.13)
regressor.fit(x_train, y_train)

trainPrediction = regressor.predict(x_train)
trainR = r2_score(y_train, trainPrediction)
trainMSE = mean_squared_error(y_train, trainPrediction)
trainMAPE = mean_absolute_percentage_error(y_train, trainPrediction)

testPrediction = regressor.predict(x_test)
testR = r2_score(y_test, testPrediction)
testMSE = mean_squared_error(y_test, testPrediction)
testMAPE = mean_absolute_percentage_error(y_test, testPrediction)

print(f"Train R: {trainR}, MSE: {trainMSE}, MAPE: {trainMAPE}")
print(f"Test R: {testR}, MSE: {testMSE}, MAPE: {testMAPE}")

print(regressor.score(x_test, y_test))


```
